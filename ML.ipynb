Encoding categorical data

```python
category_cols_new=train_data.select_dtypes(include='object').columns
train_data_new=pd.get_dummies(train_data,columns=category_cols_new,prefix=category_cols_new)
test_data_new=pd.get_dummies(test_data,columns=category_cols_new,prefix=category_cols_new)
```

separacion de train y test 

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(train_data_new.drop('SalePrice',axis=1),train_data_new['SalePrice'],test_size=0.20,random_state=42)
```

```python
x_train.shape,y_train.shape,x_test.shape,y_test.shape
```

Escaling

```python
from sklearn.preprocessing import StandardScaler
scale=StandardScaler()
x_train_scale = scale.fit_transform(x_train)
x_test_scale = scale.transform(x_test)
```

Extraccion de caracteristicas

```python
from sklearn.decomposition import PCA
sklearn_pca = PCA(n_components=0.95)
sklearn_pca.fit(x_train_scale)
```

```python
x_train_transformed = sklearn_pca.transform(x_train_scale)
x_test_transformed =sklearn_pca.transform(x_test_scale)
```

```python
x_train_transformed.shape,x_test_transformed.shape
```

Modelo de datos

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_absolute_percentage_error,mean_squared_log_error
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.linear_model import LinearRegression
from xgboost import XGBRegressor
```

```python
#Creo 5 modelos distintos
RF = RandomForestRegressor(random_state=42).fit(x_train_transformed, y_train.values)
DT = DecisionTreeRegressor(random_state=42).fit(x_train_transformed, y_train.values)
GBR = GradientBoostingRegressor(random_state=42).fit(x_train_transformed, y_train.values)
LR = LinearRegression().fit(x_train_transformed, y_train.values)
XGB = XGBRegressor().fit(x_train_transformed, y_train.values)
```

```python
# evaluo sus metricas
models = [LR, DT, RF, GBR, XGB]
RMSLE = [mean_squared_log_error(y_test.values, mod.predict(x_test_transformed)) for mod in models]
```

```python
# comparo los 5 modelos
Models = ['Linear Regression','Decision Tree','Random Forest','Gradient Boosting','XgBoost']
evaluation = pd.DataFrame({'Models':Models,'RMSLE':RMSLE})
```
